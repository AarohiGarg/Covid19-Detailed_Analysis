```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

covid_data = pd.read_csv('country_wise_latest.csv') 

```

```python
highest = covid_data.sort_values(by='Confirmed').tail(10)
highest.plot(x = 'Country/Region', y = 'Confirmed', kind = 'bar', title = '10 Countries with Maximum Confirmed Cases')
```

```python
ratio = covid_data["Confirmed"] / covid_data["Deaths"]
covid_data['Ratio'] = ratio
new_data = covid_data.loc[covid_data.Ratio != float('inf')]
new_data = new_data.sort_values(by=['Ratio']).tail(5)
new_data.plot(x = 'Country/Region', y = 'Ratio', kind = 'bar', title = '5 countries that have highest total cases to death ratio')
```

```python
ratio = covid_data["Confirmed"] / covid_data["Recovered"]
covid_data['Ratio'] = ratio
new_data = covid_data.loc[covid_data.Ratio != float('inf')]
new_data = new_data.sort_values(by=['Ratio']).tail(10)
new_data.plot(x = 'Country/Region', y = 'Ratio', kind = 'bar', title = '5 countries that have highest total cases to Recovered cases')
```

```python
lowest = covid_data.sort_values(by='Confirmed').head(10)
lowest.plot(x = 'Country/Region', y = 'Confirmed', kind = 'bar', title = '10 Countries with minimum Confirmed Cases')
```

```python
highest = covid_data.sort_values(by='Active').tail(10)
highest.plot(x = 'Country/Region', y = 'Active', kind = 'bar', title = '10 Countries with Maximum Active Cases')
```
