```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

covid_data = pd.read_csv('state_level_latest.csv') 
covid_data.drop(covid_data.columns[[5,6,7,8,9,10,11]], axis = 1, inplace = True) 
covid_data = covid_data[covid_data.State != 'Total']
covid_data
```

```python
highest = covid_data.sort_values(by='Confirmed').tail(10)
highest.plot(x = 'State', y = 'Confirmed', kind = 'bar', title = '10 States with Maximum Confirmed Cases')
```

```python
ratio = covid_data["Confirmed"] / covid_data["Deaths"]
covid_data['Ratio'] = ratio
new_data = covid_data.loc[covid_data.Ratio != float('inf')]
new_data = new_data.sort_values(by=['Ratio']).tail(5)
new_data.plot(x = 'State', y = 'Ratio', kind = 'bar', title = '5 states that have highest total cases to death ratio')
new_data
```

```python
ratio = covid_data["Confirmed"] / covid_data["Recovered"]
covid_data['Ratio'] = ratio
new_data = covid_data.loc[covid_data.Ratio != float('inf')]
new_data = new_data.sort_values(by=['Ratio']).tail(10)
new_data.plot(x = 'State', y = 'Ratio', kind = 'bar', title = '5 states that have highest total cases to Recovered cases')
```

```python
lowest = covid_data.sort_values(by='Confirmed').head(10)
lowest.plot(x = 'State', y = 'Confirmed', kind = 'bar', title = '10 States with minimum Confirmed Cases')
```

```python
highest = covid_data.sort_values(by='Active').tail(10)
highest.plot(x = 'State', y = 'Active', kind = 'bar', title = '10 States with Maximum Active Cases')
```
